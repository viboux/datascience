```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
df = pd.read_csv('C:/Dev/python/aircraft/boeingvsairbus.csv',sep=';')
```

```python
unpiv = pd.melt(df, id_vars=['Annee'], value_vars=df.columns[1:], var_name='modele', value_name='n')
unpiv['temps'] = pd.to_datetime(unpiv['Annee'], format='%Y')
unpiv.drop('Annee', 1, inplace=True) #Dropper colonne Annee (1 = col, 0 = row)
unpiv.index = unpiv['temps']
del unpiv['temps']
unpiv
```

```python
unpiv['nsum'] = unpiv.groupby(['modele'])['n'].cumsum()
unpiv['previous_nsum'] = unpiv.groupby(['modele'])['n'].transform(lambda x : x.cumsum().shift()) #apply
unpiv['rsum'] = unpiv.groupby(['modele'])['n'].transform(lambda x : x.rolling(10, min_periods=1).sum()) #apply
unpiv['previous_rsum'] = unpiv.groupby(['modele'])['n'].transform(lambda x : x.shift().rolling(10, min_periods=1).sum()) #apply
unpiv.head(50)
```

```python
#unpiv['previous_nsum'] = unpiv.nsum.shift(1)
unpiv['rank'] = unpiv.groupby('temps')['nsum'].rank(ascending=False)
unpiv['rank_rsum'] = unpiv.groupby('temps')['rsum'].rank(ascending=False)
unpiv['1975-01-01']
```

```python
output = pd.DataFrame()
output['name'] = unpiv['modele']
output['value'] = unpiv['nsum'].fillna(0) #.replace(0, 'NA')
output['year'] = unpiv.index.year
output['lastValue'] = unpiv['previous_nsum'].fillna(0) #.replace(0, 'NA')
output['rank'] = unpiv['rank'].astype('int32')
output.to_csv('dataset_d3.csv', index=False)
output
```

```python
output_rsum = pd.DataFrame()
output_rsum['name'] = unpiv['modele']
output_rsum['value'] = unpiv['rsum'].fillna(0) #.replace(0, 'NA')
output_rsum['year'] = unpiv.index.year
output_rsum['lastValue'] = unpiv['previous_rsum'].fillna(0) #.replace(0, 'NA')
output_rsum['rank'] = unpiv['rank_rsum'].astype('int32')
output_rsum.to_csv('dataset_d3_rsum.csv', index=False)
output_rsum
```
